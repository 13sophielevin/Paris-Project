```python
import glob
import re

#Function for using regex patterns for replacements.
def lookup(s, patterns):
    for pattern, value in patterns:
        if re.search(pattern, s):
            return value
    return None



path = '/home/seth/Notes/HDW/test/*.txt'
filenames = glob.glob(path)                 #Establish the list of files on which to operate.
subs = {'aicnt': 'aient', ' allerà ': ' aller à ', 'avecun': 'avec un', ' bonre ': ' bonne ', ' cct ': ' cet ', ' cctte ': ' cette ', ' cst ': ' est ', 
        ' ct ': ' et ', ' cu ': ' eu ', 'cux ': 'eux ', 'ccrtain': 'certain', ' dansla': ' dans la', 
        ' dansle': ' dans le', ' dansles': ' dans les', 'daris': 'dans', 'dù': 'dû', 'êlre':'être', 'enx': 'eux', 
        ' ge ': ' ce ', 'ilest': 'il est',
        ' lcur ': ' leur ', 'lamême': 'la même', 'lemême': 'le même', ' nc ': ' ne ', 'Panis': 'Paris', 'plusicurs': 'plusieurs', 'qne': 'que', 
        ' scra ': ' sera ', ' scrai': ' serai', ' surla ': ' sur la ', ' surle ': ' sur le ', 
        ' surles ': ' sur les ', 'T1': 'Il', ' toutc ': ' toute ', ' toutcs ': ' toutes ', '1à': 'là', '1e': 'le', 
        '1a': 'la', '1\]': 'Il', '1%': '1er', '1!': 'Il', '!a': 'la', '!e': 'le', '[*': '1er', '_': ' ' 
         }

patterns = [(r'[\w,]\n+[a-zA-Z.,?_:;]', r'[\w,]\s[a-zA-Z.,?_:;]'), (r'[\w,]\n+\s', r'[\w,]\s'), 
        (r'[\w-]\n+[a-zA-Z]', r'[\w-][a-zA-Z]')]

for name in filenames:                      #Loop over all files in corpus. For each file:
    oldname = name.rsplit('.')
    newname = oldname[0]+'_CL.txt'          #Create a new name for the cleaned file
    
    file = open(name)                       #open and read the existing file into variable "raw".
    raw = file.read()
    #print(raw)                              #Debugging only
    

    #Deletions and replacements using the dictionary happen here.
    for key, value in subs.items():
        if key in raw:
            raw = raw.replace(key, value)
    
    
    #Deletions and replacements using regex patterns.
    for key, value in patterns:
        if key in raw:
            raw = raw.replace(key, patterns(key))
        
    
    
    clean_file = open(newname, "w+")       #Write the cleaned-up string to the new file, and close it.
    clean_file.write(raw)
    clean_file.close()
    print(raw)                        #Debugging
```
