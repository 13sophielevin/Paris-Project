```python
import glob
from nltk import pos_tag

paths_to_files = glob.glob('/home/seth/Notes/HDW/test/*.txt')

print(len(paths_to_files),'files in path')
proper_nouns = []
for file in paths_to_files:
    text = open(file).read()
    #for word in text:
     #   if word[0].isupper():
      #proper_nouns.append(word[0:])
    tagged_text = pos_tag(text.split())
    for word,pos in tagged_text:
        if pos == 'NNP':
            proper_nouns.append(word)
print('done')
print(proper_nouns)        
```
